```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

SMALL_DIR   = "/kaggle/input/handwritten-words-dataset/data/small"
CAPITAL_DIR = "/kaggle/input/handwritten-words-dataset/data/Capital"

IMG_SIZE = (64, 128)
BATCH = 32
SEED = 42

# Завантажуємо обидва набори
ds_small = tf.keras.utils.image_dataset_from_directory(
    SMALL_DIR,
    image_size=IMG_SIZE,
    batch_size=BATCH,
    color_mode="rgb",
    shuffle=True,
    seed=SEED
)

ds_capital = tf.keras.utils.image_dataset_from_directory(
    CAPITAL_DIR,
    image_size=IMG_SIZE,
    batch_size=BATCH,
    color_mode="rgb",
    shuffle=True,
    seed=SEED
)

# Об'єдн
all_ds = ds_small.concatenate(ds_capital)

# Список класів
class_names = ds_small.class_names + ds_capital.class_names
num_classes = len(class_names)

print("Класи:", class_names)
print("К-сть класів:", num_classes)

# Нормалізація
all_ds = all_ds.map(lambda x,y:(x/255.0,y))
```

```python
# train / val / test
n_batches = tf.data.experimental.cardinality(all_ds).numpy()

n_train = int(0.7 * n_batches)
n_val   = int(0.15 * n_batches)

train_ds = all_ds.take(n_train)
temp_ds  = all_ds.skip(n_train)
val_ds   = temp_ds.take(n_val)
test_ds  = temp_ds.skip(n_val)

print("train:", n_train, "val:", n_val, "test:", n_batches-n_train-n_val)

num_classes = len(class_names)

# CNN
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(64, 128, 3)),

    tf.keras.layers.Conv2D(32, 3, activation="relu"),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Conv2D(64, 3, activation="relu"),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Conv2D(128, 3, activation="relu"),
    tf.keras.layers.MaxPooling2D(),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(0.3),

    tf.keras.layers.Dense(128, activation="relu"),
    tf.keras.layers.Dense(num_classes, activation="softmax")
])

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# навчання 
history = model.fit(train_ds, validation_data=val_ds, epochs=8)

test_loss, test_acc = model.evaluate(test_ds, verbose=0)
print("\ntest точність:", round(float(test_acc), 4))

images, labels = next(iter(test_ds))
pred = model.predict(images, verbose=0)
pred_ids = np.argmax(pred, axis=1)

plt.figure(figsize=(12,4))
for i in range(min(8, len(images))):
    plt.subplot(2,4,i+1)
    plt.imshow(images[i].numpy())
    t = class_names[int(labels[i])]
    p = class_names[int(pred_ids[i])]
    plt.title(f"T:{t}\nP:{p}")
    plt.axis("off")
plt.tight_layout()
plt.show()
```
